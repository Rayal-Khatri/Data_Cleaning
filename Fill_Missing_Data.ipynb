# Dependencies

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Dataset

```python
# Load dataset
dataset_path = r"https://drive.google.com/uc?export=download&id=1EYv3V2koljcYcDpsqZi7QFsmjmOwBRtI"
df = pd.read_csv(dataset_path)
df.shape
```

```python
df.head(6)
```

```python
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',None)
```

```python
df.head(6)
```

```python
df.isnull().sum()
```

```python
missing_val= df.isnull().sum()/df.shape[0] *100
missing_val
```

```python
remove_val = missing_val[missing_val>20].keys()
remove_val
```

```python
df2 = df.drop(columns = remove_val)
df2.shape
```

```python
df2 = df2.select_dtypes(include=['int64', 'float64'])
df2.head(4)
```

```python
df2.isnull().sum()
```

```python
# List of columns with missing values
columns_with_null = [var for var in df2.columns if df2[var].isnull().sum() > 0]
columns_with_null
```

```python
plt.figure(figsize=(15,15))
sns.set()

for i, var in enumerate(columns_with_null):
    plt.subplot(2,2,i+1)
    sns.histplot(
        df2[var],
        bins = 20,
        kde= True,
        # color='#DC143C',
        line_kws={'linewidth': 5},
    )
```
