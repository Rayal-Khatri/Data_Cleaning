## Checking Heatmap for missing data. 
In this case the white spots represent the places where the data is missing. The white spots needs to be removed

```python
plt.figure(figsize=(25,25))
sns.heatmap(df.isnull())
```

```python
null_var= df.isnull().sum()/df.shape[0] * 100         # checking percentage od data that is missing
null_var
```

```python
drop_columns = null_var[null_var>17].keys()
drop_columns
```

```python
df = df.drop(columns=drop_columns)
```

```python
plt.figure(figsize=(25,25))
sns.heatmap(df.isnull())
```

```python
df = df.dropna()
```

```python
df.shape
```

```python
plt.figure(figsize=(25,25))
sns.heatmap(df.isnull())
```

```python
df.isnull().sum().sum()
```

```python
num_var = df.select_dtypes(include=['int64', 'float64']).columns
num_var
```

```python
plt.figure(figsize=(16,9))
sns.histplot(
    orginal_df['MSSubClass'],
    kde= True,
    color = 'r',
    label = "Orginal Dataset"
)
sns.histplot(
    df['MSSubClass']+ .5,
    kde= True,
    color = 'g',
    label = "Cleaned Dataset"
)

plt.legend()
```

```python
plt.figure(figsize=(55,45))

for i,var in enumerate(num_var):
    plt.subplot(9,5, i+1)
    sns.histplot(
        orginal_df[var],
        bins=20,
        kde= True,
        color = 'r',
        label = "Orginal Dataset"
    )
    sns.histplot(
        df[var]+ .5,
        bins=20,
        kde= True,
        color = 'g',
        label = "Cleaned Dataset"
    )
```

```python
def cat_var(var):
    return pd.concat(
    [orginal_df[var].value_counts()/df.shape[0] *100  ,df[var].value_counts()/df.shape[0] *100],
    axis=1 ,
    keys= [var+'_org',var +'_clean']
 )
```
